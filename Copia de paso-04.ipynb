Crea la clase `Word` en el script `ApylabradosModule.py`

* Configura el constructor para que tenga el atributo `word`, que será un array de caracteres inicialmente vacío.
* Configura el método `.__str__()` para que dado un objeto de la clase `Word`, imprima la palabra como un string.
* Construye el método `.areEqual()` que toma por parámetro un objeto de la clase `Word`, `w`, y calcula si el atributo `word` del objeto de la clase `Word` es igual al atributo `word` del objeto `w` que toma por parámetro. En definitiva, construye un método que devuelva `True` si las dos palabras son iguales y `False` en caso contrario.
* Construye el método `.isEmpty()` que devuelva `True` si una palabra es vacía y `False` en caso contrario.
* Construye un método de clase llamado `.readWord()` que permita leer una palabra por teclado y devuelva la palabra convertida en un objeto de la clase `Word`.

```
class Word():

    def __init__(self, word = []):
        self.word = word

    def __str__(self):
        return "".join(self.word)
    
    def areEqual(self, w):
        return self.word == w.word
    
    def isEmpty(self):
        return len(self.word) == 0
    
    @classmethod
    def readWord(cls):
        w = input().upper()
        w = list(w)
        return cls(w)
```

```
Word(['h', 'o', 'l', 'a']).areEqual(Word(['h', 'a', 'l', 'a']))
```

```
Word(['h', 'o', 'l', 'a']).isEmpty()
```

```
Word([]).isEmpty()
```

```
Word().readWord()
```
