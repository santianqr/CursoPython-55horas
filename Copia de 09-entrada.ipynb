¡Ayuda a Pyratilla a crear una lista de tuplas, llamada `weapons`, donde en la primera posición de cada tupla esté el nombre del arma y en la segunda, el número de veces que hay que acertar al comebarcos con dicha arma para derrotarlo!

```
weapons = [ ("Flecha", 20),
            ("Espada", 5),
            ("Daga", 15),
            ("Puñal", 12),
            ("Martillo", 7),
            ("Hacha", 10),
            ("Maza", 3),
            ("Machete", 8)
            ]
```

armas.png

En total, 8 comebarcos han subido a bordo. La tripulación decide ir eliminándolos uno por uno. Cada uno con un arma diferente de las que disponen.

¡Ayuda a Pyratilla y su tripulación a eliminar los 8 comebarcos! Cada uno de los enemigos debe ser derrotado con un arma diferente de las disponibles usando solamente dos bucles `for`.

```
enemies = 8
print("Hay que derrotar {} enemigos\n".format(enemies))

la = ["Flecha", "Espada", "Daga", "Maza"]

for w, h in weapons:
    if w in la:
        art = "la"
    else:
        art = "el"
    for i in range(1, h+1):
        if i > 1:
            print("Le damos al comebarcos {} veces con {} {}".format(i, art, w.lower() ) )
        else:
            print("Le damos al comebarcos {} vez con {} {}".format( i, art, w.lower() ) ) 
    enemies -= 1
    print("\nQuedan {} enemigo/s por derrotar\n".format( enemies ) )
```

```
coins = [
         ("Alegre", "moneda", "amarillo"),
         ("Arrecife", "coralina", "azul"),
         ("Verde", "rupia", "verde"),
         ("Golosa", "chuche", "rosa"),
         ("Lejana", "lita", "rojo"),
         ("Espesura", "gema", "morado"),
         ("Torva", "corona", "naranja")
]

for isla, moneda, color in coins:
    print("En la isla {} se usa la {} de color {}".format( isla, moneda, color ) )
```
