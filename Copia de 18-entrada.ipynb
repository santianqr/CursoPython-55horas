# La última isla: Isla Lejana

Llegaron por fin a Isla Lejana. Tras desembarcar para ir en busca del cofre del tesoro de la isla, unos extraños personajes salieron a su encuentro. En total había 3, exactamente iguales, pero de diferentes tamaños.

Fardelejos.png

El de mayor tamaño habló: "Hola, extraños. Os damos la bienvenida a nuestra isla: Isla Lejana. Nosotros somos la guardia real. Venimos a comunicaros que, para poder explorar libremente nuestra isla, la ley dice que tenéis que redactar un documento `txt` en el que expliquéis el motivo de vuestra visita."

Cuando el mayor parecía haber acabado de hablar, antes de que los Pyrates pudieran hacer ninguna pregunta, el mediano dijo: "Si el rey de los Fardelejos, nosotros los habitantes de Isla Lejana, tras leer el documento acepta vuestra presencia en la isla, entonces podréis quedaros un máximo de 3 días. De lo contrario, tendréis que partir inmediatamente."

Tras unos segundos de silencio, el más pequeño en tamaño habló, con una voz sorprendentemente más grave de lo que los Pyrates habían imaginado: "El documento tiene que ser escrito en un máximo de 30 minutos. Nosotros os avisaremos cuando dicha cantidad de tiempo haya pasado. Antes de empezar, ¿tenéis alguna duda?"

"Sí, disculpa, tengo unas pocas dudas al respecto", contestó Pyratilla. El Fardelejo grande volvió a hablar: "Dinos tus dudas, extraño. Estamos aquí para resolverlas y para asegurarnos de que se cumpla la ley de nuestro rey."

"Gracias. Disculpa que todavía no me haya presentado. Soy Pyratilla, capitán de los Pyrates.", dijo Pyratilla. Siguió hablando él: "Mi primera duda es sobre la extensión del documento. ¿Hay algún mínimo o máximo que deba cumplirse?". El Fardelejo mediano fue quien contestó: "Hay un mínimo, que son 10 frases. No hay máximo, pues hay límite de tiempo. ¿Cuál es tu siguiente pregunta, Pyratilla?".

Ya habían dejado por fin de llamarlo extraño. El que lo llamaran por su nombre, Pyratilla lo interpretó como una buena señal. "Muchas gracias, señor... ¿cómo podría dirigirme a vosotros?". El Fardelejo pequeño contestó inmediatamente: "Dado que nos has proporcionado amablemente tu nombre, Pyratilla, no encuentro motivos para no decirte los nuestros. Mi nobre es Jitos, el más grande de nosotros se llama Far y el nombre del último de nosotros, pero no menos importante, es Dele. Somos la guardia real: Far, Dele y Jitos. ¿Era esa tu última duda?".

"Encantado de conoceros, Far, Dele y Jitos. Perdonadme, pero esa no era la duda a la que me refería. La segunda duda es con respecto al contenido. Habéis dicho que debemos indicar nuestros motivos de visita. ¿Implica eso hablar de todo lo anterior a la llegada de la isla para aportar contexto o simplemente basta hablar de por qué motivo hemos atracado en este puerto?"

Fue Far quien contestó a esta pregunta: "Un poco de contexto previo os haría ganar puntos en la favorable decisión de nuestro rey, pero no es obligatorio. ¿Era esta tu última pregunta?"

Pyratilla miró a sus tripulantes. Ambos negaron con la cabeza en señal de que no tenían ninguna duda adcicional. Entonces, Pyratilla dijo: "Efectivamente, esa era mi última pregunta. Creo que todo ha quedado bastante claro. Podemos empezar a contar esos 30 minutos."

"Perfecto." dijo Dele. "Os dejamos volver al barco a redactar el documento. Tras haber pasado 30 minutos Lejos os avisará y le tendréis que entregar el documento. Si consideráis haber finalizado antes de que acabe el tiempo, podéis entregárnoslo sin problema. Estaremos esperando aquí mismo."

Dicho esto, Pyratilla, Pym, Pyerce y Pyo subieron de nuevo a su barco para ponerse manos a la obra en la redacción del documento.

Una vez en el camarote de Pyratilla, el capitán dijo: "Lo primero es crear el archivo `txt`".
