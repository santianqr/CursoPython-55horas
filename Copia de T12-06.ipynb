Del dataframe `words` vamos a transformar la columna `start` a dataframe para iterar sobre sus filas.

A continuación vamos a transformar la columna `length` en lista para eliminar una de sus entradas con el método `.remove()`.

```
def is_palindrome(word): 
  """
  Devuelve si la palabra word es palíndroma.
  Args:
    word: Palabra en formato string
  Returns:
    isPalindrome: Booleano
  """
  word = word.lower()
  l = []
  isPalindrome = True
  for c in word: 
    l.append(c) 
  n = len(l)
  for i in range(int(n / 2)):
    if l[i] != l[n - (i + 1)]: 
      isPalindrome = False
  return isPalindrome

words = ["sol", "ala", "cama", "duro", "bueno", "kayak", "marea", "rotor", "misterio", "acurruca"]
data = {"word": words,
        "length": map(len, words),
        "start": map(lambda w: w[0], words),
        "end": map(lambda w: w[-1], words),
        "isPalindrome": map(is_palindrome, words)}

import pandas as pd
words = pd.DataFrame(data)
words = words.set_index("word")
words.index.names = [None]
```

```
start_df = words["start"].to_frame()
for row in start_df.itertuples():
  print("La palabra {} empieza por {}".format(row[0], row[1]))
```
