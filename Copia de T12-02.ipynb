# Ejercicio

Del dataframe `words` creado en el ejercicio anterior, vamos a mostrar cada columna por separado con un bucle `for`.

Finalmente, vamos a quedarnos con las 3 últimas columnas, y este nuevo subdataframe lo guardaremos en una nueva variable, `words2`, para no modificar el dataframe original.

```
def is_palindrome(word): 
  """
  Devuelve si la palabra word es palíndroma.
  Args:
    word: Palabra en formato string
  Returns:
    isPalindrome: Booleano
  """
  word = word.lower()
  l = []
  isPalindrome = True
  for c in word: 
    l.append(c) 
  n = len(l)
  for i in range(int(n / 2)):
    if l[i] != l[n - (i + 1)]: 
      isPalindrome = False
  return isPalindrome

words = ["sol", "ala", "cama", "duro", "bueno", "kayak", "marea", "rotor", "misterio", "acurruca"]
data = {"word": words,
        "length": map(len, words),
        "start": map(lambda w: w[0], words),
        "end": map(lambda w: w[-1], words),
        "isPalindrome": map(is_palindrome, words)}

import pandas as pd
words = pd.DataFrame(data)
words = words.set_index("word")
words.index.names = [None]
```

```
for col in list(words):
  print("=== {} ===".format(col.upper()))
  print(words[col], end = "\n\n\n")
```
