# Ejercicio



Vamos a crear una función recursiva que lleve a cabo una cuenta atrás.

```
def countdown(n):
  """
  Cuenta atrás recursiva
  Args:
    n: Número entero positivo
  Returns:
    0 cuando la cuenta atrás ha finalizado
  """
  if n == 0:
    return 0
  print(n)
  return countdown(n - 1)
```

```
countdown(10)
```
