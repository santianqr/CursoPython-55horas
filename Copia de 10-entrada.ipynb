¡Ayuda a Pyratilla a crear un método llamado `convert_to_coins` que dada una cantidad y el nombre de la moneda, realice la conversión a monedas de Isla Alegre!


```
def convert_to_coins(amount, coin_name):
    coin_name = coin_name.lower()
    equivalences = {"coralina": 0.75, "rupia":0.55, "chuche":1.30,
                    "lita": 2.15, "gema":1.70, "corona":0.95,
                    "moneda": 1}
    return equivalences[coin_name] * amount

convert_to_coins(3, "Coralina")
```

¡Ayuda a Pyratilla a aplicar dicha función para sumar al total de sus monedas, `coins`, todas las adquiridas en el botín de Isla Arrecife!

```
coins_obtained = {"moneda": 1050, "coralina": 1310, "rupia":2025,
                  "chuche":1520, "lita":970, "gema": 1975, "corona":1760}

moneda = 0
for coin in coins_obtained:
    moneda += convert_to_coins( coins_obtained[coin], coin)

coins += moneda
print(coins)
```

No obstante, Pyratilla no estaba satisfecho con solo es función y quiere crear una que convierta de cualquier tipo de moneda a cualquier tipo de moneda utilizando la función `convert_to_coins` anterior.

¡Ayuda a Pyratilla a crear esa nueva función, llamada `currency_converter`, que por parámetro reciba la cantidad, la divisa de partida y la divisa a la que quiere convertirse!

```
def convert_to_coins(amount, coin_name):
    coin_name = coin_name.lower()
    equivalences = {"coralina": 0.75, "rupia":0.55, "chuche":1.30,
                    "lita": 2.15, "gema":1.70, "corona":0.95,
                    "moneda": 1}
    return equivalences[coin_name] * amount

def currency_converter(amount, s_currency, f_currency):
    s_currency , f_currency = s_currency.lower(), f_currency.lower()
    s_currencyTo_moneda = convert_to_coins(amount, s_currency)
    equivalences = {"coralina": 0.75, "rupia":0.55, "chuche":1.30,
                    "lita": 2.15, "gema":1.70, "corona":0.95,
                    "moneda": 1}
    return s_currencyTo_moneda / equivalences[f_currency]

print(convert_to_coins(3, "coralina"))
print(currency_converter(3, "coralina", "rupia"))
```

Ayuda a Pyratilla a crear la función de la inscripción la cual debe de leer como parámetro un texto (un solo string) y como resultado debe devolver el número de palabras de longitud par y el número de palabras de longitud impar que hay en dicho texto, en una tupla donde la primera entrada sea el número de palabras pares y la segunda, el número de palabras impares. Además, deberás aplicarla al texto que se lee en la inscripción y decir cuántas palabras pares y cuántas impares hay. Solo así conseguirás abrir el cofre.

```
string = """Si quieres ver lo que hay en mi interior,
esto es lo que vas a hacer:
Crea una función llamada num_even_odd_words,
que por parámetro un texto va a leer,
y el número de palabras pares e impares de este verso
es lo que deberá correctamente devolver"""

def num_even_odd_words(string):

    string = string.lower()
    par = 0
    total = 0
    for w in string.split():
        if w.isalpha():
            if len(w) % 2 == 0:
                par += 1
                total += 1
            else:
                total += 1

    return par, total - par

par, impar = num_even_odd_words(string)
print("En el texto hay {} palabras de longitud par y {} palabras de longitud impar".format( par, impar ) )
```
