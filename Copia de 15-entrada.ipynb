# Isla Golosa


Llegaron sin inconvenientes a Isla Golosa. Desembarcaron y empezaron la búsqueda del tesoro. Todo estaba muy tranquilo... demasiado tranquilo...

Por un lado, todos estaban felices pues durante el viaje hasta Isla Golosa y todo el rato que habían pasado en la Isla no había habido ningún incidente, pero aún así, Pyratilla sentía que algo no iba bien. Todo había ido demasiado bien para ser cierto.

Absorto estaba en sus pensamientos cuando de repente Pym gritó: "¡El cofre del tesoro!". El hecho de haberlo encontrado tan rápido aumentó las sospechas de nuestro querido protagonista.

isla_golosa_cofre_cerrado.png

Se acercaron al cofre para leer la inscripción de la cerradura mágica, pero esta vez no había ningún texto. Nada que indicara cómo abrir el cofre.

Frustrado, Pyratilla quiso forzar el candado. Fue entonces cuando, al posar sus manos sobre el cofre, todo se volvió negro. Instantes después, volvió la luz y Pyratilla observó que ya no estaban en las dulces y vívidas sendas de Isla Golosa, sino que se encontraban dentro de una especie de templo, que no había visto la luz ni visitantes en muchos, muchos años.

Tras comprobar que todos sus tripulantes se encontraban bien, decidieron ponerse en marcha y encontrar una salida. Observaron la abandonada sala en la que se encontraban. Fue entonces cuando se percataron de que el cofre ya no estaba. En la alta y triste sala solamente había una puerta grande en una de las paredes de piedra y una ventana pequeña e inalcanzable en la pared opuesta. Con la luz que entraba por ese ventanuco pudieron ver que las otras paredes estaban desnudas. Por tanto, no tenían más remedio que salir por la puerta.

Al acercarse, vieron que sobre las puertas de piedra oscura, tras una capa de musgo y polvo, había letras grabadas. Tras limpiar un poco la suciedad que cubría la inscripción, pudieron leer:

<dl>
<dd>"Con ansia querías el tesoro,</dd>
<dd>ahora encontrarlo debes</dd>
<dd>si salir de aquí quieres.</dd>
<dd>Avanza y no mires atrás,</dd>
<dd>nada se te ha perdido</dd>
<dd>a lo largo del camino recorrido.</dd>
</dl>

Tras haber leído la última estrofa, abrieron las puertas y pasaron a la siguiente sala.

La nueva habitación era muy parecida a la anterior. Sin embargo, esta estaba mejor iluminada pues tenía 3 ventanas pequeñas por las cuales entraban rayos de luz. Cada una se encontraba en lo más alto de una pared. La pared que no tenía ventana, se encontraba al otro lado de la sala y tenía una gran puerta de piedra oscura. El suelo estaba dividido en cuadrados iguales. La única zona que difería del resto era sobre la cuál se encontraban los piratas Pyrates. Miraron más atentamente y pudieron leer:

<dl>
<dd>"Si pisas una baldosa cuadrada,</dd>
<dd>su problema inscrito deberás resolver.</dd>
<dd>Elige una columna de baldosas</dd>
<dd>será tu camino a recorrer</dd>
<dd>y no sabrás si lo has superado</dd>
<dd>hasta el último resultado leer.</dd>
<dd>Solo de ese modo desactivarás la trampa</dd>
<dd>y evitarás así tu vida perder.</dd>
<dd>Llega al final con luz verde</dd>
<dd>para a la siguiente sala poder acceder.</dd>
</dl>

Miraron el suelo con más atención. Se trataba de una cuadrícula con 5 filas de 5 baldosas cada una. Por tanto, concluyeron que solo deberían resolver 5 pruebas para llegar a la puerta y salir de esta habitación. Entre todos decidieron resolver las pruebas de la columna central, que era la que quedaba inmediatamente delante de ellos. Pyratilla le dijo a su tripulación que él asumiría el riesgo de activar las baldosas y que contaba con su ayuda para desactivar todas las trampas y llegar al final de la sala todos juntos. Sanos y salvos.

Y sin más dilación, Pyratilla pisó la primera baldosa cuadrada. Acto seguido, una luz blanca surgió del suelo. Un texto apareció iluminado:

> "Resuelve la siguiente operación, pero solamente toma la parte entera por exceso usando el módulo `math`"

$$9\pi + 11\tau$$


```
import math
math.ceil(9 * math.pi + 11 * math.tau)
```

Nada ocurrió, así que Pyratilla decidió continuar a la siguiente baldosa. Esta vez, la luz blanca formaba el siguiente mensaje:

> "Calcula la norma euclídea del vector 2-dimensional (7, 24) y réstale 4! usando métodos del módulo `math`".

```
math.sqrt(7**2 + 24**2) - math.factorial(4)
```

Tras unos instantes, Pyratilla dio unos pasos y llegó a la tercera baldosa. Se encontraba ya a mitad de camino y de momento seguía de una pieza. Miró al suelo para ver que el texto iluminado esta vez indicaba

> "Dada la siguiente cuadrícula, colorea las casillas cuyos números sean divisibles entre 7 o entre 3. Para ello usa el módulo `math`"

cuadricula.png

```
numbers = [[7,91,195,180,220],
           [231,110,22,53,89],
           [105,6,223,29,461],
           [35,83,61,757,383],
           [21,175,81,273,147]]

for row in numbers:
    print("")
    for e in row:
        if math.fmod(e,3) == 0 or math.fmod(e,7) == 0:
            print(e, end="\t")
```

Pyratilla, decidido, se movió a la cuarta baldosa. Esta vez, el texto iluminado decía:

> "Calcula $e^{3 + 5i}$ y a la parte real del resultado réstale la parte imaginaria y quédate con la parte entera usando los módulos `math` y `cmath` cuando sean necesarios"


```
import cmath
z = cmath.exp(3+5j)
math.floor(z.real - z.imag)
```

Por fin, Pyratilla llegó a la última baldosa y leyó:

> "Calcula, siendo $x = 2\cdot (-1.5708 + 0.42723i)$,

$$\sin(4.8x)\cos(2.3x)\tan(2.9x)$$

> y quédate con la parte entera por defecto de la parte real del resultado. Utiliza los módulos `math` y `cmath` donde corresponda.

```
x = 2 * (-1.5708 + 0.42723j)
e1 = cmath.sin(4.8*x) * cmath.cos(2.3*x) * cmath.tan(2.9*x)
math.floor(e1.real)
```

Pyratilla, feliz por estar de una sola pieza llegó a la puerta. Su felicidad se disipó cuando leyó:

<dd>y del quinto el caracter ASCII habrás de buscar.</dd>
<dd>Junta por orden las letras obtenidas</dd>
<dd>y si has acertado en todo, podrás continuar</dd>
</dl>
