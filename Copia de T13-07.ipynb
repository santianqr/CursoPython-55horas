# Ejercicio

Dada la clase `Point2D`, vamos a crear la clase `Point3D` que hereda de `Point2D`.

Vamos a modificar todos los métodos y usar el método `.super()` donde sea necesario para poder usarlos con puntos de 3 dimensiones.

```
class Point2D():
  
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __str__(self):
    return "({}, {})".format(self.x, self.y)

  @classmethod
  def zero(cls):
    return cls(0, 0)
```

```
class Point3D(Point2D):

  def __init__(self, x, y, z):
    super().__init__(x, y)
    self.z = z

  def __str__(self):
    return super().__str__()[:-1] + ", {})".format(self.z)

  @classmethod
  def zero(cls):
    return cls(0, 0, 0)
```

```
p = Point2D(1, -1)
```

```
print(p)
```

```
cero2D = Point2D.zero()
print(cero2D)
```

```
q = Point3D(1, 2, 3)
print(q)
```

```
cero3D = Point3D.zero()
print(cero3D)
```
