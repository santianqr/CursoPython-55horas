# Ejercicio

Vamos a convertir una lista de grados Celsius a grados Fahrenheit. El resultado lo mostraremos como una lista.

```
def from_C_to_F(celsius):
  """
  Convierte grados celsius en Fahrenheit
  Args:
    celsius: Número real
  Returns:
    fahrenheit: Número real
  """
  fahrenheit = (celsius * 9 / 5) + 32
  return fahrenheit

celsius = [0, 5, 10, 15, 20, 25, 50, 100, 200]
list(map(from_C_to_F, celsius))
```
